```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)
data = (1, 3, 2, 4)
axes.plot(data)
plt.show()
```

```python
figure = plt.figure()
axes1 = figure.add_subplot(2,2,1)
axes2 = figure.add_subplot(2,2,2)
axes3 = figure.add_subplot(2,2,3)
axes4 = figure.add_subplot(2,2,4)

axes1.plot((0, 2))
axes2.plot((1, 1))
axes3.plot((2, 0))
axes4.plot((1, 2))
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x = (0, 2, 4, 6)
y = (0, 4, 0, 2)
x2 = (0, 1, 2, 3, 4, 5, 6)
y2 = (1, 2, 3, 4, 5, 6, 7)
axes.plot(x, y, color = 'r', linestyle='dotted', marker='^')
axes.plot(x2, y2, color='k', linestyle='dashed', marker='o')
plt.show()
```

```python
figure = plt.figure(figsize=(3,3))
axes = figure.add_subplot(111)

label = ['a','b','c']
data = [30, 20, 50]

axes.pie(data, labels=label)
plt.show()
```
