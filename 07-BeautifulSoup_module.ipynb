```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://movie.daum.net/ranking/reservation"
response = requests.get(URL)
html = response.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
print(soup.find('a', class_ ='link_txt').text)
```

```python
movie_list = soup.find_all('a', class_='link_txt')
print(movie_list)
```

```python
for i in range(0, 10):
    print(i+1, end="")
    print("위 : " + movie_list[i].text)
```

```python
URL = "https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=100"

headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"}
response = requests.get(URL, headers=headers)
html = response.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
headline_list = soup.find_all('a', class_='sh_text_headline nclicks(cls_pol.clsart)')
print(headline_list)
```

```python
for i in range(0, 10):
    print(i+1, end="")
    print("번 헤드라인 : " + headline_list[i].text) 
```
